```python
import numpy as np
import scipy.io
import time
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import roc_auc_score,roc_curve,auc,average_precision_score,precision_recall_curve
from sklearn.metrics import confusion_matrix
```

```python
DATA_PATH = 'deepsea_train/'
```

### Loading data 

```python
## TEST DATA
test_data_raw = scipy.io.loadmat(DATA_PATH+'test.mat')
y_test = test_data_raw['testdata']
```

```python
y_pred = np.load('pred_deep_sea_model_20191022.npy')
```

```python
for i in range(0,125):
    fpr, tpr, thresholds = roc_curve(y_test[:, i], y_pred[:, i])
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=1)
    plt.xlim([0, 1.05])
    plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    
plt.title('ROC Transcription Factors')
plt.show()
```
